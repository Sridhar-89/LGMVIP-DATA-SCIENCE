Image to Pencil Sketch with Python

Dataset link:https://image.cnbcfm.com/api/v1/image/103522929-Bentley_Continental_GT_Speed_Breitling_Jet_Team_Series_Limited_Edition.jpg?v=1529471152&w=1920&h=1080

```python
import matplotlib.pyplot as plt
from skimage import io
import cv2
from google.colab.patches import cv2_imshow
```

dataset link:https://image.cnbcfm.com/api/v1/image/103522929-Bentley_Continental_GT_Speed_Breitling_Jet_Team_Series_Limited_Edition.jpg?v=1529471152&w=1920&h=1080

```python
image=io.imread("https://image.cnbcfm.com/api/v1/image/103522929-Bentley_Continental_GT_Speed_Breitling_Jet_Team_Series_Limited_Edition.jpg?v=1529471152&w=1920&h=1080")
```

```python
cv2_imshow(image)
```

```python
gray_filter = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_filter)
```

```python
inverted_filter = cv2.bitwise_not(gray_filter)
cv2_imshow(inverted_filter)
```

```python
image_smoothing = cv2.GaussianBlur(inverted_filter, (21, 21), sigmaX=0, sigmaY=0)
cv2_imshow(image_smoothing)
```

```python
def pencil_sketch(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch = pencil_sketch(gray_filter, image_smoothing)
cv2_imshow(pencil_sketch)
```
